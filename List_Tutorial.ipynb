### Introduction to Lists in Python
This notebook will cover the basics of Python lists, including:
1. Creating Lists
2. Accessing Elements
3. Modifying Lists
4. Common List Methods
5. Iterating Through Lists
6. List Comprehensions
7. Nested Lists

### Main Features of Lists
- **Ordered**: The order of elements is maintained. Elements can be accessed by their index.
- **Mutable**: Lists can be modified after creation (add, remove, or change elements).
- **Heterogeneous**: A single list can contain elements of different data types (e.g., integers, strings, objects).
- **Dynamic**: Lists can grow or shrink as needed. No need to declare size beforehand.
- **Support for Nesting**: Lists can contain other lists as elements, enabling the creation of complex data structures.


```python
# 1. Creating Lists
fruits = ["apple", "banana", "cherry"]
print("Fruits List:", fruits)

# 2. Accessing Elements
print("First fruit:", fruits[0])
print("Last fruit:", fruits[-1])

# 3. Modifying Lists
fruits[1] = "blueberry"
print("Modified List:", fruits)
fruits.append("orange")
print("After Append:", fruits)
fruits.remove("cherry")
print("After Remove:", fruits)

# 4. Common List Methods
numbers = [3, 1, 4, 1, 5, 9, 2]
print("Original Numbers:", numbers)
numbers.sort()
print("Sorted Numbers:", numbers)
numbers.reverse()
print("Reversed Numbers:", numbers)
print("Index of 5:", numbers.index(5))
numbers.insert(2, 7)
print("After Insert:", numbers)

# 5. Iterating Through Lists
for fruit in fruits:
    print("Fruit:", fruit)

# 6. List Comprehensions
squares = [x**2 for x in range(10)]
print("Squares:", squares)

# 7. Nested Lists
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]
print("Matrix:", matrix)
print("Element at (1, 2):", matrix[1][2])
```
